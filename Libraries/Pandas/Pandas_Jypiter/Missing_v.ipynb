```python
import numpy as np
import pandas as pd
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

"""Seaching missing values"""

```python
DF=pd.DataFrame(data)
DF
```

```python
DF.isna()#this will return valid as false and null as true
```

```python
DF.isna().sum()#sum all the null values in a columns
```

```python
DF.isna().any()# return true if there is any null value in the column and return false if there is none
```

```python
DF.notna()#it will return valid as true and null as false
```

"""Removing missing values"""

```python
DF.dropna()#remove all rows which contains null values
```

```python
DF.dropna(how='all')#remove those rows which contains all null values
```

```python
DF.dropna(axis=1)#remove columns which contains null value
```

```python
DF.dropna(thresh=3)#remove rows which contains N number of non-NULL values
```

""" Filling missing values"""

```python
DF.fillna(0)
```

```python
values={'A':0,'B':1,'C':2,'D':3}#filling different values in all columns
DF.fillna(value=values)
```

```python
#filling mean inplace of null values
df.fillna(df.mean())
```

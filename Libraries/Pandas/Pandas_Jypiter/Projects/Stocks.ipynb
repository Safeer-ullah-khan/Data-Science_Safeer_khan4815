```python
dividend_rows["Dividend Amount"] = (#creating new column dividend amount
    dividend_rows["Close"]#selecting column close in dividend rows
    .str.extract(r'(\d+\.?\d*)')#extracting the number from the text
    .astype(float)#convert text into float
)
```

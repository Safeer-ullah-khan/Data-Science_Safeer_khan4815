```python
import numpy as np
import pandas as pd
```

"""Merging"""

```python
Employee=pd.DataFrame({
    'Employee_id':[12,43,45,56,76],
    'Employee_name':['safeer','khan','fahad','tahir','saad_cholla'],
    'Department':['Data Science','Data Analyst','Wela','lull stack developer','washroom']
})
Employee
```

```python
Salaries=pd.DataFrame({
    'Employee_id':[12,43,45,56,76,54,98],
    'Salary':[20000,40000,60000,80000,70000,30000,100000],
    'Bonus':[10000,23000,45000,12000,34000,6000,70000]
})
Salaries
```

```python
pd.merge(Employee,Salaries,on='Employee_id',how='inner')#keeps the intersection of the keys e.g (Employee_id)
```

```python
pd.merge(Employee,Salaries,on='Employee_id',how='outer')#doing the union of the all rows in both dataframes
```

```python
pd.merge(Employee,Salaries,on='Employee_id',how='left')#left means merge on the basis of left dataframes eg like employee
```

```python
pd.merge(Employee,Salaries,on='Employee_id',how='right')#right means merge on the basis of right dataframe eg like salaries
```

"""Concatination"""

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
#concatinating the two dataframes in columns 
pd.concat([df1,df2])# we will always provide a list to the concatinate function
```

```python
#concatinate the two dataframes in rowsp
pd.concat([df1,df2],axis=1)
```

"""Joining"""

```python
Frame1=pd.DataFrame({
    "name":["bob","Alice","Charlie"]
},index=[1,2,3])
```

```python
Frame2=pd.DataFrame({
    "Score":[87,98,85]
},index=[2,3,4])
```

```python
Frame1.join(Frame2)
```

```python
Frame2.join(Frame1,how="outer")
```

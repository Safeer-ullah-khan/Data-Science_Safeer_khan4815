```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

"""Matplotlib Basics"""

```python
x=np.linspace(0,5,11)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y)
plt.title("safeer")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

"""SubPlots in matplotlib"""

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,x)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x**3,y)
```

"""object oriented way"""

```python
fig=plt.figure()
axis1=fig.add_axes([0.1,0.1,0.7,0.9])#First two values are for dragging the chart whiel other for hieght and width of a chartbb
axis1.plot(x,y)
axis2=fig.add_axes([0.2,0.5,0.2,0.3])
axis2.plot(x,y)
axis1.set_title("First")
axis2.set_title("Second")
```

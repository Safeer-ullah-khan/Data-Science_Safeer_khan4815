```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

"""DISTRIBUTION PLOTS"""
Distribution plots are used to visualize the quantitative data.
Some plots that allows us to visualize the distruibution of a dataset. those are following:
1.Displot/Histplot
2.joint plot
3.pai plot
4.rugplot

```python
df=sns.load_dataset('tips')
df
```

```python
"""Columns can be numerical or categorical based on the column data.
If data is in categorical form than it will be categorical columns e.g like smoker and time etc.
categorical columns can also have numerical values if numbers make categorical sense """
df.dtypes
```

```python
df['size'].unique()
```

```python
#Distribution plot (Histogram) also we can use histplot
#(kde)kernal density estimator
sns.displot(df['total_bill'],bins=20,kde=True)
```

```python
sns.displot(df['total_bill'],kind="kde")
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],kde=True)
```

```python
#jointplot
sns.jointplot(data=df, x='total_bill',y='tip')
```

```python
sns.jointplot(data=df, x='total_bill',y='tip',kind="hex")
```

```python
#pairplot
sns.pairplot(df, hue="time",kind="reg")
```

```python
sns.pairplot(df,vars=['tip','total_bill'],hue='size',palette="bright")
```

```python
#rugplot
sns.rugplot(df,palette="plasma")
```

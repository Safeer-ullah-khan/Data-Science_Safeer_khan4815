## 📊 Problem 1: E-commerce Sales Analysis

You work for an online store. Here's the sales data for 4 products over 5 days:

Each row = one product, Each column = one day

```python
sales_data = np.array([[120, 135, 148, 162, 175],  # Product A
                       [98, 105, 110, 115, 120],   # Product B
                       [200, 195, 210, 205, 220],  # Product C
                       [150, 160, 155, 170, 165]]) # Product D

print("Sales Data (Products × Days):")
print(sales_data)
```

### Task 1a: Find total sales for each product (across all 5 days)

```python
# Your code here
total_per_product = np.sum(sales_data,axis=1)

print("Total sales per product:", total_per_product)
```

### Task 1b: Find total sales for each day (across all products)

```python
# Your code here
total_per_day = np.sum(sales_data,axis=0)

print("Total sales per day:", total_per_day)
```

### Task 1c: Which product had the highest average daily sales?

```python
# Your code here
avg_sales = np.mean(sales_data,axis=1)
best_product = np.argmax(avg_sales)

print("Average sales per product:", avg_sales)
print("Best performing product index:", best_product)
```

### Task 1d: Find all sales numbers greater than 150

```python
# Your code here
high_sales = sales_data[sales_data>150]

print("Sales above 150:", high_sales)
```

## 🎓 Problem 2: Student Performance Analysis

You're a teacher analyzing test scores. You have scores for 6 students across 4 exams.

```python
# Each row = one student, Each column = one exam
exam_scores = np.array([[85, 92, 78, 88],  # Student 1
                        [76, 81, 85, 79],  # Student 2
                        [92, 95, 90, 94],  # Student 3
                        [68, 72, 75, 70],  # Student 4
                        [88, 85, 91, 87],  # Student 5
                        [95, 98, 96, 99]]) # Student 6

print("Exam Scores:")
print(exam_scores)
```

### Task 2a: Calculate each student's average score

```python
# Your code here
student_averages = np.mean(exam_scores,axis=1)

print("Each student's average:", student_averages)
```

### Task 2b: Which exam was the hardest? (lowest average score)

```python
# Your code here
exam_averages = np.mean(exam_scores,axis=0)
hardest_exam = np.argmin(exam_averages)

print("Average score per exam:", exam_averages)
print("Hardest exam (index):", hardest_exam)
```

### Task 2c: How many students scored above 85 on average?

```python
# Your code here (hint: use boolean indexing and count)
count_above_85 = np.sum(exam_averages>85)

print("Number of students with avg > 85:", count_above_85)
```

### Task 2d: Give a 5-point bonus to all scores below 80

```python
# Your code here (be careful not to modify the original!)
adjusted_scores = exam_scores.copy()  # Make a copy first
adjusted_scores[adjusted_scores<80] +=5

print("Adjusted scores:")
print(adjusted_scores)
```

## 💰 Problem 3: Monthly Expenses Tracker

Track your monthly expenses across 3 months in different categories.

```python
# Each row = category, Each column = month
expenses = np.array([[1200, 1250, 1300],  # Rent
                     [400, 450, 380],     # Food
                     [150, 200, 180],     # Transport
                     [80, 90, 100],       # Entertainment
                     [200, 180, 220]])    # Utilities

categories = ['Rent', 'Food', 'Transport', 'Entertainment', 'Utilities']
print("Monthly Expenses:")
print(expenses)
```

### Task 3a: What's the total spending per month?

```python
# Your code here
monthly_total = np.sum(expenses,axis=0)

print("Total per month:", monthly_total)
```

### Task 3b: What's the average spending per category over 3 months?

```python
# Your code here
category_avg = np.mean(expenses,axis=1)

for i, cat in enumerate(categories):
    print(f"{cat}: ${category_avg[i]:.2f}")
```

### Task 3c: What percentage of total expenses is Rent?

```python
# Your code here
total_expenses = np.sum(expenses)
rent_total = np.sum(expenses[0])
rent_percentage = (rent_total/total_expenses)*100

print(f"Rent is {rent_percentage:.1f}% of total expenses")
```

## 🏃 Problem 4: Fitness Tracker - Heart Rate Data

You recorded your heart rate (BPM) during a 7-day workout plan.

```python
# Heart rate measurements (BPM) for 7 days
heart_rate = np.array([72, 85, 78, 92, 88, 95, 80])

print("Heart rate (BPM) for 7 days:", heart_rate)
```

### Task 4a: What's your average heart rate?

```python
# Your code here
avg_hr = np.mean(heart_rate)

print(f"Average heart rate: {avg_hr:.1f} BPM")
```

### Task 4b: On which days did your heart rate exceed 85 BPM?

```python
# Your code here (hint: use np.where())
high_hr_days = np.where(heart_rate==85)

print("Days with HR > 85:", high_hr_days)
```

### Task 4c: What's the difference between max and min heart rate?

```python
# Your code here
hr_range = np.max(heart_rate)-np.min(heart_rate)

print(f"Heart rate range: {hr_range} BPM")
```

## 🌡️ Problem 5: Weather Data Analysis

Temperature data (°C) for 5 cities over one week.

```python
# Each row = city, Each column = day of week
temperatures = np.array([[28, 30, 29, 31, 32, 30, 28],  # City A
                        [15, 17, 16, 18, 19, 17, 16],  # City B
                        [22, 24, 23, 25, 26, 24, 23],  # City C
                        [35, 36, 37, 38, 37, 36, 35],  # City D
                        [10, 12, 11, 13, 14, 12, 11]]) # City E

cities = ['City A', 'City B', 'City C', 'City D', 'City E']
print("Temperature data (°C):")
print(temperatures)
```

### Task 5a: Which city has the highest average temperature?

```python
# Your code here
avg_temps = np.mean(temperatures,axis=1)
hottest_city_index = np.argmax(avg_temps)

print("Average temperatures:", avg_temps)
print(f"Hottest city: {cities[hottest_city_index]}")
```

### Task 5b: What was the coldest day across all cities?

```python
# Your code here
coldest_temp = np.min(temperatures)

print(f"Coldest temperature recorded: {coldest_temp}°C")
```

### Task 5c: Convert all temperatures to Fahrenheit

```python
# Your code here
temps_fahrenheit = (temperatures *9/5) + 32

print("Temperatures in Fahrenheit:")
print(temps_fahrenheit)
```

### Task 5d: How many temperature readings were above 25°C?

```python
# Your code here
count_above_25 = np.sum(temperatures>25)

print(f"Number of readings above 25°C: {count_above_25}")
```

## 🎮 Problem 6: Gaming Statistics

You play 3 different games and track your scores over 5 sessions.

```python
# Each row = game, Each column = session
game_scores = np.array([[1200, 1450, 1350, 1500, 1600],  # Game 1
                       [850, 900, 920, 950, 980],       # Game 2
                       [2100, 2200, 2050, 2300, 2400]]) # Game 3

print("Game Scores:")
print(game_scores)
```

### Task 6a: What's your highest score in each game?

```python
# Your code here
high_scores = np.max(game_scores, axis=1)

print("Highest scores per game:", high_scores)
```

### Task 6b: Calculate the improvement from first to last session for each game

```python
# Your code here
improvement = game_scores[:,-1]-game_scores[:,0]

print("Score improvement per game:", improvement)
```

### Task 6c: Which game shows the most consistent performance? (lowest std deviation)

```python
# Your code here
std_devs = np.std(game_scores, axis=1)
most_consistent = np.argmin(std_devs)

print("Standard deviation per game:", std_devs)
print(f"Most consistent game (index): {most_consistent}")
```

## 📱 Problem 8: Social Media Engagement

Track likes, comments, and shares for your 5 posts.

```python
# Each row = [likes, comments, shares]
engagement = np.array([[120, 15, 8],   # Post 1
                      [450, 32, 25],  # Post 2
                      [89, 7, 3],     # Post 3
                      [310, 28, 18],  # Post 4
                      [520, 45, 35]]) # Post 5

print("Engagement [likes, comments, shares]:")
print(engagement)
```

### Task 8a: What's the total engagement (sum of all metrics) for each post?

```python
# Your code here
total_engagement = np.sum(engagement)

print("Total engagement per post:", total_engagement)
```

### Task 8b: Which post got the most likes?

```python
# Your code here (hint: likes are in column 0)
most_liked_post = np.argmax(engagement[:,0])

print(f"Post with most likes: Post {most_liked_post + 1}")
```

### Task 8c: What's the average number of comments across all posts?

```python
# Your code here (comments are in column 1)
avg_comments = np.mean(engagement[:,1])

print(f"Average comments: {avg_comments:.1f}")
```

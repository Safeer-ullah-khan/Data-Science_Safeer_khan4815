```python
import numpy as np
```

```python
#Stacking of vectors
arr1=np.array([1,2,3,4,5])
arr2=np.array([6,7,8,9,10])
```

```python
np.hstack((arr1,arr2))
```

```python
np.vstack((arr1,arr2))
```

```python
np.column_stack((arr1,arr2))
```

```python
#spliting arrays
arr=np.arange(1,17).reshape(4,4)
```

```python
arr
```

```python
np.hsplit(arr,2)
```

```python
a=np.vsplit(arr,4)
a
```

```python
for i in a:
    print(i)
```

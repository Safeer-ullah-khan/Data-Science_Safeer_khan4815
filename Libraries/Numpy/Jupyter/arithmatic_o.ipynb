```python
#Deep and shallow copy 
arr=np.arange(1,21)
arr
```

```python
#deep copy
#when you change sliced part of array it does not change orignal array because the sliced part is stored at seperated memory location
Slice=arr[:5]
Slice
```

```python
Slice *2
```

```python
arr
```

```python
#shallow copy
#when you change sliced part in deep copy ,it will change the orignal array because this sliced part are stored in same memory location
b=arr.view()
b[0:2]=[23,34]
```

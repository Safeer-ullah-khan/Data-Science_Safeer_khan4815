```python
import numpy as np
```

```python
arr=np.array([20, 21, 22, 22, 23, 24, 25, 26, 200])
```

```python
Q1=np.percentile(arr,25)
Q1
```

```python
Q3=np.percentile(arr,75)
Q3
```

```python
IQR=Q3-Q1
```

```python
UF=Q3 +(1.5 * IQR)
```

```python
LF=Q1 - (1.5 * IQR)
```

```python
updated_list=[]
for i in arr:
    if i >=LF and i<=UF:
        updated_list.append(i)
```

```python
arr2=np.array(updated_list)
```

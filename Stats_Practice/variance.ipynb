```python
import numpy as np
```

```python
arr=np.array([85,90,95,100,105])
arr
```

```python
average=np.mean(arr)
average
```

```python
Com_differ=(85-average)**2 + (90-average)**2 +(95-average)**2 + (100-95)**2 +(105-95)**2
Com_differ
```

```python
count=np.size(arr)
count
```

```python
variance=Com_differ/count
variance
```

```python
import math
```

```python
STD=math.sqrt(variance)
STD
```
